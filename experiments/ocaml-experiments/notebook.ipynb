```python
import numpy as np 
import pandas as pd 
import json
import os 
import matplotlib.pyplot as plt
```

```python
PARSED_FILE = './parsed.json'
with open(PARSED_FILE, 'r') as json_file:
    data = json.load(json_file)

# Create a DataFrame from the list of dictionaries
df = pd.json_normalize(data)
df
```

```python
rbt = df[df['workload'] == 'RBT']
print("RBT, Type-Based Generator:")
display(rbt[rbt['strategy'] == 'typeBasedGenerator'].describe())
print("RBT, Bespoke Generator:")
display(rbt[rbt['strategy'] == 'bespokeGenerator'].describe())
```

```python
stlc = df[df['workload'] == 'STLC']
print("stlc, Type-Based Generator:")
display(stlc[stlc['strategy'] == 'typeBasedGenerator'].describe())
print("stlc, Bespoke Generator:")
display(stlc[stlc['strategy'] == 'bespokeGenerator'].describe())
```
